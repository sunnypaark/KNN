```python
from keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test

import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5): 
    print(y_test[i],'\n') # Print the label
    print(X_test[i],'\n') # Print the matrix of values
    plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
    plt.show()
```

```python
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of X_train[0]:", X_train[0].shape)
print("Shape of X_test[0]:", X_test[0].shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

# 3.

```python
N_train = X_train.shape[0]  
X_train_reshaped = np.array([X_train[i].reshape((1, 784)) for i in range(N_train)])
X_train_reshaped = X_train_reshaped.reshape((N_train, 784))

N_test = X_test.shape[0]  
X_test_reshaped = np.array([X_test[i].reshape((1, 784)) for i in range(N_test)])
X_test_reshaped = X_test_reshaped.reshape((N_test, 784))

print(X_train_reshaped.shape)
print(X_test_reshaped.shape)
```

## 4.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

k_values = [1, 3, 5, 7, 9]
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_reshaped, y_train)
    y_pred = knn.predict(X_test_reshaped)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    print(f"k={k}, Accuracy: {acc:.4f}")

best_k = k_values[np.argmax(accuracies)]
print("Best k:", best_k)
```
